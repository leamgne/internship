```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np
import scipy.stats as stats
```

```python
contingency_table = pd.crosstab(df['HPV result'], df['categories '])

# Perform the Chi-Square test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Output the test results
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
```

```python
chi2, p, _, _ = chi2_contingency(contingency_table)

# Calculate Cramer's V as an effect size measure
n = contingency_table.values.sum()
phi_c = np.sqrt(chi2 / (n * min(contingency_table.shape) - 1))

# Output the results
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Cramer's V: {phi_c}")
```

```python
import numpy as np

def cramers_v(confusion_matrix):
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

# Calculate Cramer's V
cramer_v = cramers_v(contingency_table.values)
```
